```python
documents = [
    "Python es un lenguaje de programación muy usado en inteligencia artificial.",
    "FAISS es una biblioteca creada por Facebook para búsqueda vectorial eficiente.",
    "Los embeddings convierten texto en vectores numéricos.",
    "RAG significa Retrieval Augmented Generation.",
    "LangChain permite construir aplicaciones con modelos de lenguaje.",
    "Kubernetes es una plataforma de orquestación de contenedores que permite automatizar el despliegue, escalado y gestión de aplicaciones.",
    "Docker es una herramienta que permite crear, ejecutar y administrar contenedores para facilitar el desarrollo y despliegue de aplicaciones.",
    "Linux es un sistema operativo de código abierto ampliamente utilizado en servidores, desarrollo y sistemas de red.",
    "Las redes de computadoras permiten la comunicación entre dispositivos para compartir información y recursos."
]
```

```python
from langchain_text_splitters import CharacterTextSplitter

# Configuración del splitter
text_splitter = CharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=0
)

docs = text_splitter.create_documents(documents)

# Probar salida
for doc in docs:
    print(doc.page_content)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(docs, embeddings)

print("Vectorstore creado con éxito")
```

```python
query = "¿Qué son los kubernetes?"

results = vectorstore.similarity_search(query)

for r in results:
    print(r.page_content)
```
